# Resistor circuits

## Overview

We will now begin to analysis circuits, computing various physical quantities for the elements in the circuit. So far, you have only been introduced to batteries and resistors, so our circuits will be constructed only from those devices in this lesson. Later, in Lesson 33 (capacitor circuits), we will add in another element type, the capacitor.

There are a few ways to go about circuit analysis. One method, which we will talk more about in Lesson 32 (complex circuits), is to use Kirchhoff's laws to write down a set of equations mathematically describing the relationships between the currents and voltages in the circuit. That is a powerful technique, but sometimes it is more than is necessary.

In this lesson, we will use the method of simplification. We take a group of resistors, and find a single equivalent resistor that has the exact same electrical properties as the original group; then, we can "replace" the group with the equivalent resistor, leading to a simpler circuit. If we do this over and over, eventually a circuit with a single resistor -- equivalent to all the resistors in the original circuit -- and the battery. At this point, we can use Ohm's law, since that relation only works for a single resistor. Undoing the process, we can calculate the properties of the original, physical resistors. This will be the scheme we use in this lesson.

## Series circuits

To illustrate the simplification method above, let's start with a circuit with a group of resistors all in series, i.e.  the circuit only has a single branch, with no junctions. We want to find a single resistor that has the same electric properties as the group, so how can we do this?

![Replacing a group of series resistors with a single equivalent resistor](../img/INT_AY21_L31_Fig01_Series_resistors.png)

By the junction rule, all of the resistors will have the same circuit, since they all lie on the same branch. Thus, $I_1 = I_2 = \cdots = I$. Also, the total voltage of the group is simply the sum of the individual voltages, so $V_{tot} = V_1 + V_2 + \cdots$. We call the resistance of the equivalence resistor $R_{eq}$, and use Ohm's law to set it equal to $R_{eq} = V_{tot} / I$. Remember that each of the physical resistors will have, for example, $R_1 = V_1 / I_1$. Let's now start with Ohm's law for the equivalent resistor and see if we can write it in terms of the individual resistors.

$$
    R_{eq} = \frac{V_{tot}}{I} = \frac{V_1 + V_2 + \cdots}{I} = \frac{V_1}{I_1} + \frac{V_2}{I_2} + \cdots = R_1 + R_2 + \cdots
$$

Here, we used the fact that the currents are the same. This shows us that we can replace the group of series resistors with a single equivalent resistor, whose resistance is merely the sum of the individual resistors. This gives the equivalent resistance for a group of series resistors.

> **Series equivalent resistance:** $$R_{eq} = R_1 + R_2 + \cdots$$

Now, let's see how this works in practice with an example circuit, given below. We will find the current through, the voltage across, and the power emitted, by all of the resistors in the circuit.

![An example series circuit](../img/INT_AY20_MP3_L12_Fig01_Series_circuit.png)

In order to compute the properties of the circuit, we use what I call a **VIRP table**. The name comes from the fact that we use it to find the voltage, current, and power of the resistor group. I will show the various steps to fill in the table. At this point, we can only write down the given resistances, and the total voltage (which is that of the battery). Notice that, in the upper left corner, we have put the type of grouping, to remind us of what rules to use when calculating the rest of the table!

| Series     | V (V) | I (A) | R ($\Omega$)  | P(W) |
|------------|-------|-------|---------------|------|
| $R_1$      |       |       | 10.00         |      |
| $R_2$      |       |       | 15.00         |      |
| $R_3$      |       |       | 20.00         |      |
| $R_{123}$  | 90.00 |       |               |      |

A couple of things to note:

1. We do not need to write the units in the table entries, because we have put them by the appropriate quantity at the top of each column.
1. **More importantly**, we have used *four* sig figs for each number. Get into this habit! It will not be relevant here, but when we get into more complicated circuits, using four sig figs will maintain the accuracy of your calculations.

We can group the three resistors together into a single equivalent resistor, so we redraw the circuit again, but now with the three physical resistors replaced by the equivalent resistor. This equivalent resistance is labeled as $R_{123}$, to remind us it is a combination of the resistances $R_1, R_2,$ and $R_3$. Since the group is in series,

$$
    R_{123} = R_1 + R_2 + R_3
$$

![Simplifying the series circuit](../img/INT_AY20_MP3_L12_Fig02_Series_simplify.png)

We can fill in this equivalent resistance into our VIRP table.

| Series     | V (V) | I (A) | R ($\Omega$)  | P(W) |
|------------|-------|-------|---------------|------|
| $R_1$      |       |       | 10.00         |      |
| $R_2$      |       |       | 15.00         |      |
| $R_3$      |       |       | 20.00         |      |
| $R_{123}$  | 90.00 |       | 45.00         |      |

For the bottom row, we now have enough information to find the current (using Ohm's law $R = V/I$) and the power (with $P = IV$).

| Series     | V (V) | I (A) | R ($\Omega$)  | P(W)  |
|------------|-------|-------|---------------|-------|
| $R_1$      |       |       | 10.00         |       |
| $R_2$      |       |       | 15.00         |       |
| $R_3$      |       |       | 20.00         |       |
| $R_{123}$  | 90.00 | 2.000 | 45.00         | 180.0 |

Since the physical resistors are in series, the currents are the same. We can copy the total current into the individual currents for each resistor in the table.

| Series     | V (V) | I (A) | R ($\Omega$)  | P(W)  |
|------------|-------|-------|---------------|-------|
| $R_1$      |       | 2.000 | 10.00         |       |
| $R_2$      |       | 2.000 | 15.00         |       |
| $R_3$      |       | 2.000 | 20.00         |       |
| $R_{123}$  | 90.00 | 2.000 | 45.00         | 180.0 |

Now we can finish the table, since we have enough for each resistor to use Ohm's law and the power equation.

| Series     | V (V) | I (A) | R ($\Omega$)  | P(W)  |
|------------|-------|-------|---------------|-------|
| $R_1$      | 20.00 | 2.000 | 10.00         | 40.00 |
| $R_2$      | 30.00 | 2.000 | 15.00         | 60.00 |
| $R_3$      | 40.00 | 2.000 | 20.00         | 80.00 |
| $R_{123}$  | 90.00 | 2.000 | 45.00         | 180.0 |

Although this was a relatively simple table, let's go through a couple of checks to make sure we did everything right.

1. Since the group is in series, there is an extra fact that we have not used yet: the individual voltages must add up to the total voltage.
1. In addition, for *every* group of resistors, the individual powers must add up to the total power. This is just good old fashioned conservation of energy!

## Parallel circuits

Now that we have worked through series resistor circuits, we need to do the same for parallel circuits. First, we will find the equivalent resistance of a group of parallel resistors, then work through an example of a VIRP table for such a group.

![Replacing a group of parallel resistors with an equivalent resistor](../img/INT_AY21_L31_Fig02_Parallel_resistors.png)

As shown in the figure above, we want to replace the group of parallel resistors on the left with a single resistor, as seen on the right. What value of $R_{eq}$ will allow us to do this? Remember that for parallel resistors, it is now the voltage that is the same -- we can see this using the loop rule, taking a loop passing through any pair of resistors in the group. Thus, $V_1 = V_2 = \cdots = V$. On the flip side, the individual currents all add up to the total, so $I = I_1 + I_2 + \cdots$. We will start with this second fact, and then use Ohm's law for each individual resistor. In particular, we want the total current $I$ to be related to the total voltage $V$ and the equivalent resistance $R_{eq}$ by $I = V/R_{eq}$. This means

$$
    \frac{V}{R_{eq}} = I = I_1 + I_2 + \cdots = \frac{V_1}{R_1} + \frac{V_2}{R_2} + \cdots
$$

since Ohm's law holds for each individual resistor as well. Since $V = V_1 = V_2 = \cdots$, then

$$
    \frac{V}{R_{eq}} = V \biggl( \frac{1}{R_1} + \frac{1}{R_2} + \cdots \biggr)
$$

Dividing out the common factor of voltage $V$ gives us the equivalent resistance for a group of parallel resistors.

> **Parallel equivalent resistance:** $$ \frac{1}{R_{eq}} = \frac{1}{R_1} + \frac{1}{R_2} + \cdots$$

I will warn you now -- a mistake students sometimes make is to forget to take the reciprocal of $R_{eq}$ at the end of this calculation! You must first add up the sums of the reciprocals of the individual resistance, then take the reciprocal of this sum to find the equivalent resistance. Hopefully it is obvious that the *series* equivalent resistance is always bigger than any of the individual resistances; you can show that the *parallel* equivalent resistance is always *smaller* than any of the individual resistances as well.

Let's take the same resistors and battery we used for the series example above, and place them into a parallel resistor circuit. As we work through this example, notice how the values we compute are different than those we found earlier.

![An example parallel circuit](../img/INT_AY20_MP3_L12_Fig04_Parallel_example.png)

Because the values are the same as before, our VIRP table looks similar, just changing the group type to "parallel".

| Parallel   | V (V) | I (A) | R ($\Omega$)  | P(W) |
|------------|-------|-------|---------------|------|
| $R_1$      |       |       | 10.00         |      |
| $R_2$      |       |       | 15.00         |      |
| $R_3$      |       |       | 20.00         |      |
| $R_{123}$  | 90.00 |       |               |      |

We now simplify the circuit into a single equivalent resistor $R_{123}$.

![Simplifying the parallel circuit](../img/INT_AY20_MP3_L12_Fig05_Parallel_simplify.png)

Solving for the equivalent resistance $R_{123}$,

$$
    \frac{1}{R_{123}} = \frac{1}{R_1} + \frac{1}{R_2} + \frac{1}{R_3}
$$

we can fill this into the table, then solve the bottom row using Ohm's law and the power equation. Notice that the equivalent resistance is not a round number, so keeping four sig figs is becoming important!

| Parallel   | V (V) | I (A) | R ($\Omega$)  | P(W) |
|------------|-------|-------|---------------|------|
| $R_1$      |       |       | 10.00         |      |
| $R_2$      |       |       | 15.00         |      |
| $R_3$      |       |       | 20.00         |      |
| $R_{123}$  | 90.00 | 19.50 | 4.615         | 1755 |

This is a parallel group, so now the *voltage* for the individual resistors is the same as the total voltage. Copy these up the voltage column, then again use Ohm's law and the power equation to complete the table. You should get the following results.

| Parallel   | V (V) | I (A) | R ($\Omega$)  | P(W)  |
|------------|-------|-------|---------------|-------|
| $R_1$      | 90.00 | 9.000 | 10.00         | 810.0 |
| $R_2$      | 90.00 | 6.000 | 15.00         | 540.0 |
| $R_3$      | 90.00 | 4.500 | 20.00         | 405.0 |
| $R_{123}$  | 90.00 | 19.50 | 4.615         | 1755  |

To check the accuracy of the table, we make sure (1) the individual currents add up to the total, and (2) the individual powers add up to the total. Since both of these work, the table is correct.

Even though we used the same resistors and battery, the results we get are very different! In particular, the circuit has a much larger power emission -- the energy storage of the battery is being drained a lot quicker. This is because the current is almost ten times larger. The much lower resistance of the parallel circuit, compared to the series circuit, has a profound effect on the electrical properties of the system.

## Hybrid circuits

So far, we have taken a single group of resistors, either in series or parallel, and replaced them with a single eqiuivalent resistor. However, there is no reason we cannot keep doing this process over and over. This will allow us to treat more involved circuits. It *does* depend on the ability to group resistors together in series or parallel. If we cannot do this, we will need to use a more sophisticated method, which we will see in Lesson 32 when we use Kirchhoff's laws. I will refer to circuits were you *can* group all the resistors repeatedly into series or parallel groups as a **hybrid circuit**.

As an example, let's consider the circuit diagram shown below.

![An example hybrid circuit](../img/INT_AY20_MP3_L13_Fig01_Hybrid_example.png)

The first step we need to take is to describe the circuit -- what resistors can we group together as either series or parallel? Going back to the definition of branches and junctions from Lesson 30 may be helpful here. For example, the four resistors are *not* all in series, because they are not all on the same branch. There are, in fact, three branches and two junctions. The junctions are where the wires meet, above and below the battery, so the branches are the left-hand wire, the middle wire, and the right-hand wire. This means that the resistors on the same branch are in series. Thus, $R_1$ and $R_2$ are in series, and $R_3$ is in series with $R_4$.

So now what? Well, these two branches both connect at the top and bottom junctions, so the branches form a complete loop. This means that the *group* $R_1$ and $R_2$ is in parallel with the *group* $R_3$ and $R_4$. This spells out the following roadmap for the circuit:

1. $R_1$ and $R_2$ are in series
1. $R_3$ and $R_4$ are in series
1. $R_{12}$ and $R_{34}$ are in parallel

For the last line, I have written the groupings in terms of their equivalence resistance, as a hint that we cannot do anything there until we first complete the series groupings $R_1$ and $R_2$, and $R_3$ and $R_4$. In other words, we have to start with the first two simplifications before we go further. Writing out a roadmap like this can help decide which simplifications to start with.

Now we go through the process of circuit simplification, which I spell out here. Notice that the steps listed below are basically the ones we did in the previous examples, but there are some extra things thrown in.

1. Draw the circuit diagram for the original circuit. You can rearrange the drawing, if necessary, to help you understand the circuit.
1. Make a roadmap of which resistors are in series or parallel with each other (including equivalent resistances as appropriate).
1. For every simplification, redraw the circuit with the new equivalent resistor, and set up the VIRP table matching the simplification. Find the equivalent resistance of the group in question, and fill it into the table.
1. Continue this process until you have a circuit with one battery and one resistor equivalent to all the physical resistors.
1. Work back in the reverse direction, filling in the remaining parts of the VIRP tables. *Completely* fill in each table, and check your work, before moving on to the next table! This means verifying that the powers add up to the total, for example.

Let's go through this process with the example circuit given in the figure above. I randomly picked five values for the resistances and the battery voltage. Why? The beauty of this process is that you can always write down new problems yourself, and make sure you have the right answer, by using the checks I mentioned before (Do the powers add up? Do the currents add up in parallel, or the voltages in series?). I highly recommend that you do this yourself -- make up a hybrid circuit, and values for the circuit elements, then analyze the circuit -- as a practice method.

Anyway, back to the example. I have redrawn the circuit below in a different way, to emphasize which resistors are in series or parallel.

![The redrawn hybrid example circuit](../img/INT_AY20_MP3_L13_Fig02_Hybrid_redraw.png)

For my random values, I have chosen $R_1 = 3.14 \ \Omega, R_2 = 1.59 \ \Omega, R_3 = 2.65 \ \Omega, R_4 = 3.58 \ \Omega$, and $V = 9.79$ V. We now start at the beginning of our roadmap from before, and group $R_1$ and $R_2$ together in series. At this point, we can only find the equivalent resistance $R_{12}$.

![Grouping $R_1$ and $R_2$ together in series](../img/INT_AY20_MP3_L13_Fig03_First_redraw.png)

| Series     | V (V) | I (A) | R ($\Omega$)  | P(W) |
|------------|-------|-------|---------------|------|
| $R_1$      |       |       | 3.140         |      |
| $R_2$      |       |       | 1.590         |      |
| $R_{12}$   |       |       | 4.730         |      |

> **What we can fill in:** Notice that we do *not* know the voltage yet! Unlike the series and parallel examples before, the equivalent resistor $R_{12}$ is only part of the circuit. So the voltage across is it *not* the same as the battery voltage! I often see students make this mistake, and get wildly wrong answers -- if they had checked their power column, they would know some is drastically wrong. We will not fill in the battery voltage until we have simplified the circuit completely, so we have a few more steps first.

Next, we do the same thing for the $R_3$ and $R_4$ resistors. This is independent of our previous table, so we could have done them in the opposite order, if we wanted.

![Grouping $R_3$ and $R_4$ together in series](../img/INT_AY20_MP3_L13_Fig04_Second_redraw.png)

| Series     | V (V) | I (A) | R ($\Omega$)  | P(W) |
|------------|-------|-------|---------------|------|
| $R_3$      |       |       | 2.650         |      |
| $R_4$      |       |       | 3.580         |      |
| $R_{34}$   |       |       | 6.230         |      |

Our last simplification takes the two equivalent resistors from the previous steps, and combines them in parallel. Notice that in the VIRP table, we fill in the equivalence resistances we obtained from the last two tables. We could not have done this step before we did the last two.

![Grouping $R_{12}$ and $R_{34}$ together in series](../img/INT_AY20_MP3_L13_Fig05_Final_redraw.png)

| Parallel   | V (V) | I (A) | R ($\Omega$)  | P(W) |
|------------|-------|-------|---------------|------|
| $R_{12}$   |       |       | 4.730         |      |
| $R_{34}$   |       |       | 6.230         |      |
| $R_{1234}$ | 9.790 |       | 2.689         |      |

At this point, we *can* fill in the battery voltage, since we have reduced the circuit down to one battery and one (equivalent) resistor. This is the only point we can use the total battery voltage. It is *also* the only point we can start using Ohm's law as well, since that equation only applies for a resistor with a known voltage or current.

As we did before, we can now fill in all the entries of this third VIRP table by (1) solving the bottom line with Ohm's law and the power equation, (2) use the fact that this is a parallel grouping to fill in the voltages, (3) solve the remaining rows as we did the last. This gives

| Parallel   | V (V) | I (A) | R ($\Omega$)  | P(W)  |
|------------|-------|-------|---------------|-------|
| $R_{12}$   | 9.790 | 2.070 | 4.730         | 20.27 |
| $R_{34}$   | 9.790 | 1.571 | 6.230         | 15.38 |
| $R_{1234}$ | 9.790 | 3.641 | 2.689         | 35.65 |

Verify that the currents and the powers add up to the total, to check the table is correct. Once this is done, we can copy the information for $R_{34}$ into the second table, and go through the same process. The final table for this group is

| Series     | V (V) | I (A) | R ($\Omega$)  | P(W)  |
|------------|-------|-------|---------------|-------|
| $R_3$      | 4.166 | 1.572 | 2.650         | 6.549 |
| $R_4$      | 5.628 | 1.572 | 3.580         | 8.847 |
| $R_{34}$   | 9.790 | 1.572 | 6.230         | 15.38 |

It is this table where we start seeing the reason for using four sig figs in all of our calculations. Notice that, when we add up the voltages $V_3$ and $V_4$, we get 9.794 V instead of 9.790 V. Similarly, adding up the powers $P_3$ and $P_4$, we get 15.40 W instead of 15.38 W. However, to *three* sig figs (which is what the original values were in), the currents and the powers both agree. *Going to the extra sig fig helps us maintain our accuracy in the tables*. The more steps we do, the greater the chance of rounding the numbers increasing the error.

For our final table, we copy the values for $R_{12}$ from the parallel VIRP table into the first series table. This gives

| Series     | V (V) | I (A) | R ($\Omega$)  | P(W)  |
|------------|-------|-------|---------------|-------|
| $R_1$      | 6.500 | 2.070 | 3.140         | 13.46 |
| $R_2$      | 3.291 | 2.070 | 1.590         | 6.812 |
| $R_{12}$   | 9.790 | 2.070 | 4.730         | 20.27 |

Again, to *three* sig figs, the voltages and the powers add up to the totals.

## Summary

In this lesson, we developed the idea of replacing a group of resistors with a single equivalent resistor with the same properties. This simplified the circuit, and repeating this process allows us to eventually get to a circuit we can completely solve using Ohm's law and the power equation. Reversing the simplifications then gives the properties of the physical resistors.

This process of simplification can be applied in many other contexts. If you take an electrical engineering course, you will see that a large number of such methods have been developed for other situations, such as the $Y-\Delta$ transform, and the more general star-mesh transform, for resistors. There are other such techniques when other types of circuit element are involved. All of them depend on the basic concept of replacing one group with another, simpler group with the same electrical properties.

The circuit simplification method used in this lesson only works if we can merge all resistors together using series or parallel groupings. Often this is not possible, and another path is necessary. This will be explored in the next lesson, when we return to Kirchhoff's laws. This will give a method of writing down equations relating the properties of the resistors, which can be solved for their values. Although it can be more mathematically involved, this technique is far more powerful.

* Construct the appropriate VIRP table(s)  to solve a resistor circuit.
